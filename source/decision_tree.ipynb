```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


import seaborn as sns
import random

# import standard scalaer and PCA
from sklearn.utils import shuffle
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
# import random forest classifier
from sklearn.ensemble import RandomForestClassifier
```

```python
# Import gene_expression_data
gene_expression_data = pd.read_csv('../data/gene_expression_data.csv', index_col=0)
```

```python
# applu log2 transformation to gene_expression_data
gene_expression_data = np.log2(gene_expression_data)
```

```python
gene_expression_data = gene_expression_data.T


# add column named state with progressing or stable if row contains _P or _S
for index, row in gene_expression_data.iterrows():
    if '_P' in index:
        gene_expression_data.loc[index, 'state'] = 1
    elif '_S' in index:
        gene_expression_data.loc[index, 'state'] = 0
```

```python
gene_expression_data
```

```python

# set random seed
random.seed(120)

# shuffle rows of gene_expression_data
gene_expression_data = shuffle(gene_expression_data)


#train_data.shape, test_data.shape
train_data = gene_expression_data.iloc[:220, :]
valid_data = gene_expression_data.iloc[220:280, :]
# Rest to the test data
test_data = gene_expression_data.iloc[300:, :]

#shape
# count state in the three datasets


# create a random forest classifier
rf = RandomForestClassifier(n_estimators=20, random_state=4, max_depth=200, criterion='gini', bootstrap=True)

# fit the model
rf.fit(train_data.drop('state', axis=1), train_data['state'])

train_data['state'].value_counts(), valid_data['state'].value_counts(), test_data['state'].value_counts()

```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
```

# Validation 

```python
rf.score(train_data.drop('state', axis=1), train_data['state'])

# validate the model
predictions = rf.predict(valid_data.drop('state', axis=1))

# get accuracy, f1 score and recall
accuracy = accuracy_score(valid_data['state'], predictions)
recall = recall_score(valid_data['state'], predictions)
f1 = f1_score(valid_data['state'], predictions)

print(f'Accuracy: {accuracy}')
print(f'Recall: {recall}')
print(f'F1 Score: {f1}')

```
